```python
import numpy as np
# Below are a quick examples

# Example 1: Create numpy arrays
arr = np.array([3, 5, 7, 9, 11, 15, 18, 22])

# Example 2: Use slicing a 1D arrays
arr2 = arr[1:6]
print(arr2)
# Example 3: Slice Starting from 3rd value to end
arr2 = arr[3:]
print(arr2)
# Example 4: Slice 0 to 4 index
arr2 = arr[:5]
print(arr2)
# Example 5: Use negative slicing
arr2 = arr[-4:-2]
print(arr2)
# Example 6: Use step value
arr2 = arr[::3]
print(arr2)
# Example 7: Use slicing with interval
arr2 = arr[3::4]
print(arr2)
# Example 8: Use slicing a 2D arrays
arr = np.array([[3, 5, 7, 9, 11],
               [2, 4, 6, 8, 10]])
arr2 = arr[1:,1:3]
print(arr2)
# Example 9: Slicing 3-D arrays
arr = np.array([[[3, 5, 7, 9, 11],
                 [2, 4, 6, 8, 10]],
                [[5, 7, 8, 9, 2],
                 [7, 2, 3, 6, 7]]])
arr2 = arr[0,1,0:2]
print(arr2)
```

```python
import numpy as np
arr = np.array([30,60,90])

val_sin = np.sin(arr)
print("Sine value",val_sin)

val_cos = np.cos(arr)
print("Cosine value",val_cos)

val_tan = np.tan(arr)
print("Tangent value",val_tan)
```

```python
import numpy as np
arr = np.array([30,60,90])
 
rad = np.deg2rad(arr)
print("Radian values for the array having degree values:", rad)
 
arr_rad = np.array([0.52359878, 1.04719755, 1.57079633])
degree = np.rad2deg(arr_rad)
print("Degree values for the array having radian values:", degree)
```

```python
import numpy as np
arr = np.array([1,0.5])
 
sin_ang = np.arcsin(arr)
print("Angle from the sin function:", sin_ang)
 
cos_ang = np.arccos(arr)
print("Angle from the cos function:", cos_ang)
 
tan_ang = np.arctan(arr)
print("Angle from the tan function:", tan_ang)
```

```python
import numpy as np
arr = np.array([30,60,90])
 
val_sin = np.sinh(arr)
print("Hyperbolic Sine value",val_sin)
 
val_cos = np.cosh(arr)
print("Hyperbolic Cosine value",val_cos)
 
val_tan = np.tanh(arr)
print("Hyperbolic Tangent value",val_tan)
```
